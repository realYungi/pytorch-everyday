```python
"""
Creating basic arrays automatically

"""
print(np.zeros(3))

print(np.ones(4))

print(np.arange(6))

print(np.arange(1, 10, 3))
```

```python
# spacing linearly between a range
print(np.linspace(0, 10, num=5))
```

```python
# in numpy, the default dtype is float64, so

x = np.ones(2, dtype=np.int64)
print(x)
```

## Adding, Removing, and sorting elements

```python
arr = np.array([2, 1, 5, 3, 7, 4, 6, 8])

print(np.sort(arr))
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print(np.concatenate((a, b)))
```

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])
```

```python
np.concatenate((x, y))
```

## Shape / Size of an Array

```python
array_example = np.array([[[0, 1, 2, 3],
                           [4, 5, 6, 7]],

                          [[0, 1, 2, 3],
                           [4, 5, 6, 7]],

                          [[0 ,1 ,2, 3],
                           [4, 5, 6, 7]]])
```

```python
print(array_example)
```

```python
array_example.ndim
```

```python
array_example.size
```

```python
array_example.shape
```

```python
a = np.array(
    [
        [1,2,3],
        [4,5,6]
    ]
)

print(np.shape(a))
```

```python
"""
We can also reshape an array

"""

a = np.arange(6)
b = a.reshape(3, 2)
print(a)

print(b)
```
